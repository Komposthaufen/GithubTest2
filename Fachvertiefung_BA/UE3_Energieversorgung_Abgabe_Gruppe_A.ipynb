# 1.a - Dauerlinie der Last und der Residuallast

```python
data_preis_last = pd.read_excel('Daten_Preise_Last_2012.xlsx')
data_last = pd.read_excel('Last_PV.xlsx')

# Drop the first row from the 'data_last' DataFrame
#data_last = data_last.drop(0)

last = data_last.loc[ : , "Last"]
PV_0 = data_last.loc[ : , "PV"]

last_res_0 = last - (PV_0 * 0)
last_res_50 = last - (PV_0 * 50000)
last_res_100 = last - (PV_0 * 100000)
last_res_150 = last - (PV_0 * 150000)
last_res_200 = last - (PV_0 * 200000)

#Sortieren

last_res_0_sorted_idx = np.argsort(last_res_0)[::-1]
last_res_0_sorted = last_res_0[last_res_0_sorted_idx]

last_res_50_sorted_idx = np.argsort(last_res_50)[::-1]
last_res_50_sorted = last_res_50[last_res_50_sorted_idx]

last_res_100_sorted_idx = np.argsort(last_res_100)[::-1]
last_res_100_sorted = last_res_100[last_res_100_sorted_idx]

last_res_150_sorted_idx = np.argsort(last_res_150)[::-1]
last_res_150_sorted = last_res_150[last_res_150_sorted_idx]

last_res_200_sorted_idx = np.argsort(last_res_200)[::-1]
last_res_200_sorted = last_res_200[last_res_200_sorted_idx]

arr = np.arange(8760)

plt.plot(arr, last_res_0_sorted, label ='0')
plt.plot(arr, last_res_50_sorted, label ='50')
plt.plot(arr, last_res_100_sorted, label ='100')
plt.plot(arr, last_res_150_sorted, label ='150')
plt.plot(arr, last_res_200_sorted, label ='200')

plt.legend()
plt.grid()


# Display the plot
plt.show()
```

# 1.b - Leistungsdauerkennlinie aus realen Daten

```python
# Read the data from the Excel file
data_preis_last = pd.read_excel('Daten_Preise_Last_2012.xlsx')

# Extract the data columns
last = data_preis_last.iloc[:, 4]
pv = data_preis_last.iloc[:, 6]
wind = data_preis_last.iloc[:, 5]


# Calculate the residual load
RL = last - pv - wind

# Sort the residual load data in descending order
RL_sorted = sorted(RL, reverse=True)
last_sorted = sorted(last, reverse=True)

# Create a numpy array for the x-axis values
x = range(1, len(RL_sorted) + 1)

# Create the plot
plt.plot(x, RL_sorted, label ='Residual_load')
plt.plot(x, last_sorted, label ='Load')

# Add labels and title
plt.xlabel('Hours')
plt.ylabel('Residual Load (MW)')
plt.title('Load Duration Curve for Residual Load in 2012')
plt.legend()
plt.grid()


# Show the plot
plt.show()
```

# 1.c - Scatterplots

```python
# Read the data from the Excel file
data_preis_last = pd.read_excel('Daten_Preise_Last_2012.xlsx')

# Extract the data columns
last = data_preis_last.iloc[:, 4]
pv = data_preis_last.iloc[:, 6]
wind = data_preis_last.iloc[:, 5]
price = data_preis_last.iloc[:, 7]

# Create the scatter plot for Last vs. Price
plt.scatter(last, price, 2)
plt.xlabel('Last (MW)')
plt.ylabel('Spotmarktpreis (EUR/MWh)')
plt.title('Spotmarktpreis vs. Last in 2012')
plt.show()

# Create the scatter plot for Residuallast vs. Price
RL = last - pv - wind
plt.scatter(RL, price, 2)
plt.xlabel('Residuallast (MW)')
plt.ylabel('Spotmarktpreis (EUR/MWh)')
plt.title('Spotmarktpreis vs. Residuallast in 2012')
plt.show()

# Create the scatter plot for Renewable Energy Einspeisung vs. Price
renewable = pv + wind
plt.scatter(renewable, price, 2)
plt.xlabel('Erneuerbare Einspeisung (MW)')
plt.ylabel('Spotmarktpreis (EUR/MWh)')
plt.title('Spotmarktpreis vs. Erneuerbare Einspeisung in 2012')
plt.show()
```

# 2.b - Mittleren stÃ¼ndlichen GroÃŸhandelsstrompreise

```python
# Load the data from the Excel file
data = pd.read_excel('Spotpreis.xlsx')

# Extract the two columns from the data
data_2008 = data.iloc[:, 0]
data_2016 = data.iloc[:, 8]

# Calculate the mean hourly prices for each year
mean_2008 = [data_2008[i::24].mean() for i in range(24)]
mean_2016 = [data_2016[i::24].mean() for i in range(24)]

# Create a numpy array for the x-axis values
x = range(1, len(mean_2008) + 1)

# Create the plot
plt.plot(x, mean_2008, label='2008')
plt.plot(x, mean_2016, label='2016')
plt.title('Mittleren stÃ¼ndlichen GroÃŸhandelsstrompreise')
plt.xlabel('Stunden')
plt.ylabel('Preis (EUR/MWh)')
plt.xticks(x)
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

Gut ersichtlich sind die zwei preislichen Hochpunkte (Morgens und Abends).  Der erste Peak scheint jedoch in 2016 weit besser abgefangen zu werden, was auf einen Ausbau von Solarproduktion, bessere Netzte fÃ¼r Ausgleich aus anderen LÃ¤ndern, oder einen Ausbau von SpeicherkapazitÃ¤ten hindeutet. Alternativ waren 2016 auch Gaskraftwerke durch das damals noch billige Gas aus Russland eine Option, diese wÃ¤ren zu den preislichen Peakzeiten sehr lukrativ und ein Zubau dieser kÃ¶nnte ebenfalls dazu fÃ¼hren den Strompreis auf einem niedrigeren Niveau zu halten.

# 3.a - MonetÃ¤ren Ertrag am Spotmarkt

```python
# Daten Laden
data = pd.read_excel('Spotpreis.xlsx')

# Daten Auslesen
data_2008 = data.iloc[:, 0]
data_2009 = data.iloc[:, 1]
data_2010 = data.iloc[:, 2]
data_2011 = data.iloc[:, 3]
data_2012 = data.iloc[:, 4]
data_2013 = data.iloc[:, 5]
data_2014 = data.iloc[:, 6]
data_2015 = data.iloc[:, 7]
data_2016 = data.iloc[:, 8]
data_2021 = data.iloc[:, 9]
data_2021 = pd.to_numeric(data.iloc[:, 9], errors='coerce')

# Replace NaN values with 0
data_2021.fillna(0, inplace=True)

# Load the PV generation profile
pv_profile = pd.read_excel('Last_PV.xlsx')
pv_profile = pv_profile.iloc[:, 1]

# Calculate the total energy generated by the PV system in MWh
pv_energy_2008 = np.sum(pv_profile * data_2008) / 1000
pv_energy_2009 = np.sum(pv_profile * data_2009) / 1000
pv_energy_2010 = np.sum(pv_profile * data_2010) / 1000
pv_energy_2011 = np.sum(pv_profile * data_2011) / 1000
pv_energy_2012 = np.sum(pv_profile * data_2012) / 1000
pv_energy_2013 = np.sum(pv_profile * data_2013) / 1000
pv_energy_2014 = np.sum(pv_profile * data_2014) / 1000
pv_energy_2015 = np.sum(pv_profile * data_2015) / 1000
pv_energy_2016 = np.sum(pv_profile * data_2016) / 1000
pv_energy_2021 = np.sum(pv_profile * data_2021) / 1000

# Print the results
print('Ertrag 1 kWp PV system in 2008: {:.2f} EUR'.format(pv_energy_2008))
print('Ertrag 1 kWp PV system in 2009: {:.2f} EUR'.format(pv_energy_2009))
print('Ertrag 1 kWp PV system in 2010: {:.2f} EUR'.format(pv_energy_2010))
print('Ertrag 1 kWp PV system in 2011: {:.2f} EUR'.format(pv_energy_2011))
print('Ertrag 1 kWp PV system in 2012: {:.2f} EUR'.format(pv_energy_2012))
print('Ertrag 1 kWp PV system in 2013: {:.2f} EUR'.format(pv_energy_2013))
print('Ertrag 1 kWp PV system in 2014: {:.2f} EUR'.format(pv_energy_2014))
print('Ertrag 1 kWp PV system in 2015: {:.2f} EUR'.format(pv_energy_2015))
print('Ertrag 1 kWp PV system in 2016: {:.2f} EUR'.format(pv_energy_2016))
print('Ertrag 1 kWp PV system in 2021: {:.2f} EUR'.format(pv_energy_2021))

```

# 3.b - Tage 4 bis 34 und 180 bis 220 aus den Jahren 2008, 2016 und 2021


```python
# Daten Laden
data = pd.read_excel('Spotpreis.xlsx')
data_last = pd.read_excel('Last_PV.xlsx')

# Daten Auslesen
data_2008 = data.iloc[:, 0]
data_2016 = data.iloc[:, 8]
data_2021 = data.iloc[:, 9]
data_2021 = pd.to_numeric(data.iloc[:, 9], errors='coerce')
data_last_PV = data_last.loc[ : , "PV"]

# Replace NaN values with 0 kann minimal fÃ¼r Aabweichungen sorgen
data_2021.fillna(0, inplace=True)

# Berechnen
Ertrag_2008 = (data_last_PV * data_2008 *1/1000 )
Ertrag_2016 = (data_last_PV * data_2016 *1/1000 )
Ertrag_2021 = (data_last_PV * data_2021 *1/1000 )

# MonetÃ¤re ErtrÃ¤ge (1kWp-Anlage) fÃ¼r Tage 4 bis 34 & Tage 180 bis 220 im Jahr 2008
monetary_yield_2008 = np.sum(Ertrag_2008[72:792])  # Stundenindex von Tag 4 bis 34 (3*24 : 33*24)
monetary_yield_2008 = round(monetary_yield_2008, 2)
print("MonetÃ¤re ErtrÃ¤ge (1kWp-Anlage) fÃ¼r Tage 4 bis 34 im Jahr 2008:   ", monetary_yield_2008)
monetary_yield_2008_additional = np.sum(Ertrag_2008[4296:5256])  # Stundenindex von Tag 180 bis 220 
monetary_yield_2008_additional = round(monetary_yield_2008_additional, 2)
print("MonetÃ¤re ErtrÃ¤ge (1kWp-Anlage) fÃ¼r Tage 180 bis 220 im Jahr 2008:", monetary_yield_2008_additional)

# MonetÃ¤re ErtrÃ¤ge (1kWp-Anlage) fÃ¼r Tage 4 bis 34 & Tage 180 bis 220 im Jahr 2016
monetary_yield_2016 = np.sum(Ertrag_2016[72:792])  # Stundenindex von Tag 4 bis 34
monetary_yield_2016 = round(monetary_yield_2016, 2)
print("MonetÃ¤re ErtrÃ¤ge (1kWp-Anlage) fÃ¼r Tage 4 bis 34 im Jahr 2016:   ", monetary_yield_2016)
monetary_yield_2016_additional = np.sum(Ertrag_2016[4296:5256])  # Stundenindex von Tag 180 bis 220
monetary_yield_2016_additional = round(monetary_yield_2016_additional, 2)
print("MonetÃ¤re ErtrÃ¤ge (1kWp-Anlage) fÃ¼r Tage 180 bis 220 im Jahr 2016:", monetary_yield_2016_additional)

# MonetÃ¤re ErtrÃ¤ge (1kWp-Anlage) fÃ¼r Tage 4 bis 34 & Tage 180 bis 220 im Jahr 2021
monetary_yield_2021 = np.sum(Ertrag_2021[72:792])  # Stundenindex von Tag 4 bis 34
monetary_yield_2021 = round(monetary_yield_2021, 2)
print("MonetÃ¤re ErtrÃ¤ge (1kWp-Anlage) fÃ¼r Tage 4 bis 34 im Jahr 2021:   ", monetary_yield_2021)
monetary_yield_2021_additional = np.sum(Ertrag_2021[4296:5256])  # Stundenindex von Tag 180 bis 220
monetary_yield_2021_additional = round(monetary_yield_2021_additional, 2)
print("MonetÃ¤re ErtrÃ¤ge (1kWp-Anlage) fÃ¼r Tage 180 bis 220 im Jahr 2021:", monetary_yield_2021_additional)

```
